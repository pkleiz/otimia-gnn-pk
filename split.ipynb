```python
import pandas as pd
import numpy as np
```

```python
import json
import random

# ====== CONFIGURAÇÃO ======
INPUT_PATH = "./data/new-generated/EUC_2D.json"      # seu arquivo original
TRAIN_PATH = "./data/new-generated/euc_2d_train.json"
VAL_PATH   = "./data/new-generated/euc_2d_val.json"
TEST_PATH  = "./data/new-generated/euc_2d_test.json"


TRAIN_RATIO = 0.8
VAL_RATIO   = 0.1
TEST_RATIO  = 0.1
SEED = 42
# ==========================

# Garantir que os ratios somam 1.0
assert abs((TRAIN_RATIO + VAL_RATIO + TEST_RATIO) - 1.0) < 1e-8, \
    "As razões devem somar 1.0!"

# 1. Load dataset
with open(INPUT_PATH, "r") as f:
    data = json.load(f)

print(f"Total de instâncias: {len(data)}")

# 2. Shuffle
random.seed(SEED)
random.shuffle(data)

total = len(data)
n_train = int(total * TRAIN_RATIO)
n_val   = int(total * VAL_RATIO)
# test = restante

train_set = data[:n_train]
val_set   = data[n_train : n_train + n_val]
test_set  = data[n_train + n_val :]

# 3. Save files
with open(TRAIN_PATH, "w") as f:
    json.dump(train_set, f, indent=2)

with open(VAL_PATH, "w") as f:
    json.dump(val_set, f, indent=2)

with open(TEST_PATH, "w") as f:
    json.dump(test_set, f, indent=2)

# 4. Print stats
print("=== Split final ===")
print(f"Train: {len(train_set)}")
print(f"Val:   {len(val_set)}")
print(f"Test:  {len(test_set)}")

print("\nArquivos salvos:")
print(TRAIN_PATH, VAL_PATH, TEST_PATH)
```
